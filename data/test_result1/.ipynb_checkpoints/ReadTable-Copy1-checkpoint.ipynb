```python
import numpy as np
import pandas as pd
import re
import os
import pymysql
import jieba
from collections import Counter

pd.set_option('display.max_rows',None)
```

```python
data_raw = list(pd.read_table('all.txt', header = None)[0])
```

```python
reg = r'^{"in_node": "(.*?)", "out_node": ["]?(.*?)["]?, "msg": "(.*?)"}$'
reg = re.compile(reg)
```

```python
data = []
for i in data_raw:
    data.append(list(reg.findall(i)[0]))
```

```python
data
```

```python
data_combine = dict()
for i in data:
    index = 'in: '+i[0]+' out: '+i[1]
    if index in data_combine:
        data_combine[index] += (' ' + i[2])
    else:
        data_combine[index] = i[2]
```

```python
index = list(data_combine.keys())
```

```python
index
```

```python
data_combine_statistic = dict()
for i in index:
    cut = jieba.cut(data_combine[i])
    times = dict(Counter(cut))
    sort = sorted(times.items(),key=lambda item:item[1],reverse=True)
    sort = [i for i in sort if i[0] != ' ']
    data_combine_statistic[i] = sort
```

```python
data_combine_statistic_df = []
for i in data_combine_statistic:
    tmp = pd.DataFrame(data_combine_statistic[i])
    tmp.name= i
    tmp.columns = ['word', 'count']
    data_combine_statistic_df.append(tmp)
```
